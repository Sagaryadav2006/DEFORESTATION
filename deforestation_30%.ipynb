```python
import pandas as pd
```

```python
df = pd.read_csv('deforestation.csv')
```

```python
df
```

```python
def clean_col_names(df):
    cols = df.columns
    new_cols = []
    for col in cols:
        new_col = col.replace(' ', '_').replace('(%)','percent').replace('(°)', 'deg').replace('(°C)','celcius').replace('(m)','m').replace('(7d)','7d').replace('(1=Yes,0=No)','bool').replace('(per_km²)','per_km2').lower()
        new_cols.append(new_col)
    df.columns = new_cols
    return df
df = clean_col_names(df)
print(df.columns)
```

```python
df['date'] = pd.to_datetime(df['date'])
print(df['date'])
```

```python
categorical_cols = ['region','country','biome_type','dataset_split']
for col in categorical_cols:
    print(df[col].unique())
```
